```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("full-data.csv")
df
```

```python
clipPerID=df.groupby(["clipName","pptID"]).max().reset_index()[["clipName","pptID","condition"]]
clipPerID
```

```python
questions=pd.unique(df["question"])
```

```python
clipIDQuestionResp=df.groupby(["clipName","pptID","question"]).sum().reset_index()[["clipName","pptID","question","response"]]
clipIDQuestionResp[clipIDQuestionResp["question"]=="age"]["response"]

```

```python
for q in questions:
    if q not in ["age", "openBox"]: # responses for 'age' are rather random. Let's get rid of that
        clipPerID[q] = clipIDQuestionResp[clipIDQuestionResp["question"]==q].reset_index()["response"]
    
    
```

```python
clipPerID.to_csv("questionnaire-answers-per-clip.csv")
clipPerID
```
